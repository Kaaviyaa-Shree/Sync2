# **Feature Extraction for Entire Training Dataset**

```python
import os
import librosa
import numpy as np

X = []
y = []

train_path = 'train'

dialect_folders = os.listdir(train_path)

for dialect in dialect_folders:
    dialect_path = os.path.join(train_path, dialect)

    if not os.path.isdir(dialect_path):
        continue

    for speaker_folder in os.listdir(dialect_path):
        speaker_path = os.path.join(dialect_path, speaker_folder)

        if not os.path.isdir(speaker_path):
            continue

        for file in os.listdir(speaker_path):
            if file.endswith('.wav'):
                audio_path = os.path.join(speaker_path, file)

                try:
                    audio, sr = librosa.load(audio_path, sr=16000)
                    mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)
                    mfcc_mean = np.mean(mfcc, axis=1)

                    X.append(mfcc_mean)
                    y.append(dialect)

                except Exception as e:
                    print("Error with file:", audio_path)

X = np.array(X)
y = np.array(y)

print("Feature shape:", X.shape)
print("Labels shape:", y.shape)
```

# **Label Encoding Dialect Classes**

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("Dialect classes:", le.classes_)
```

# **Train–Validation Split**

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)
```

# **Model 1 – Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

# **Feature Extraction for Test Dataset**

```python
import os
import librosa
import numpy as np

test_path = 'test'

test_files = []
test_features = []

for file in os.listdir(test_path):
    if file.endswith('.wav'):
        audio_path = os.path.join(test_path, file)

        try:
            audio, sr = librosa.load(audio_path, sr=16000)
            mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)
            mfcc_mean = np.mean(mfcc, axis=1)

            test_features.append(mfcc_mean)
            test_files.append(file)

        except Exception as e:
            print("Error with file:", file)

test_features = np.array(test_features)

print("Test feature shape:", test_features.shape)
```

# **Model 1 – Test Prediction**

```python
test_pred_encoded = model.predict(test_features)
test_pred_labels = le.inverse_transform(test_pred_encoded)

print(test_pred_labels[:10])  # just to check
```

# **Model 2 - Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)
rf_model.fit(X_train, y_train)
from sklearn.metrics import accuracy_score, classification_report

y_pred_rf = rf_model.predict(X_val)

print("Random Forest Accuracy:", accuracy_score(y_val, y_pred_rf))
print(classification_report(y_val, y_pred_rf))
test_predictions_rf = rf_model.predict(test_features)
decoded_predictions_run2 = le.inverse_transform(test_predictions_rf)
with open("Sync2_Classification_Run2.txt", "w") as f:
    for file, label in zip(test_files, decoded_predictions_run2):
        file = file.replace(".wav", "")
        label=label.replace('_Dialect','')
        f.write(f"{file} {label}\n")
```
